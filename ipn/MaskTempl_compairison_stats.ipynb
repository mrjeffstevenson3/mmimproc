```python
import sys, os
sys.path.append(os.getenv('DRSNDIR'))
from drsn.statistics.descriptive import *
pd.set_option('display.max_colwidth', 90)
pd.set_option('display.max_columns', None)
pd.set_option('display.min_rows', 200)
pd.set_option('display.max_rows', 220)
pd.set_option('display.width', 1000)
pd.set_option("display.precision", 3)
pd.options.display.float_format = "{:,.3f}".format
```

```python
orig2_4_med = comp_df.dropna()[['orig_just2_median', 'orig_just4_median']].groupby(['subjects'], axis=0).median()
new2_4_med = comp_df.dropna()[['new_just2_median', 'new_just4_median']].groupby(['subjects'], axis=0).median()
orig2_4_max = comp_df.dropna()[['orig_just2_max', 'orig_just4_max']].groupby(['subjects'], axis=0).max()
new2_4_max = comp_df.dropna()[['new_just2_max', 'new_just4_max']].groupby(['subjects'], axis=0).max()
diff2_4 = pd.DataFrame()
diff2_4 = diff2_4.assign(median_diff2 = new2_4_med['new_just2_median'] - orig2_4_med['orig_just2_median'], \
                                 median_diff4 = new2_4_med['new_just4_median'] - orig2_4_med['orig_just4_median'])
diff2_4 = diff2_4.assign(max_diff2 = new2_4_max['new_just2_max'] - orig2_4_max['orig_just2_max'], \
                                 max_diff4 = new2_4_max['new_just4_max'] - orig2_4_max['orig_just4_max'])
diff2_4[['median_diff2','median_diff4', 'max_diff2', 'max_diff4']].sort_values(by='median_diff4', ascending=False)
```

```python
median2_4_descrip = comp_df.dropna()[['orig_just2_median', 'orig_just4_median', 'new_just2_median', 'new_just4_median']].groupby(['subjects'], axis=0).describe()
median2_4_descrip[[('new_just4_median', '50%')]].sort_values(by=('new_just4_median', '50%'), ascending=False).describe()
```
